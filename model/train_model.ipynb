```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[features], df['label'], test_size=0.3, random_state=1)
```

```python
import xgboost as xgb
```

```python
params = {
        'eta': 0.08, #best 0.08
        'max_depth': 6,
        'objective': 'binary:logistic',
        'eval_metric': 'auc',
        'seed': 2,
        'gamma':0.1,
        'silent': False,
        'scale_pos_weight': 11.333333333333334
    }
```

```python
watchlist = [
    (xgb.DMatrix(X_train, y_train), 'train'),
    (xgb.DMatrix(X_test, y_test), 'test')
]
model = xgb.train(params, xgb.DMatrix(X_train, y_train), 500,  watchlist, maximize=False, verbose_eval=5, early_stopping_rounds=50)
```
